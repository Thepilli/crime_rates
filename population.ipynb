```python
import pandas as pd
import os

# Replace 'your_file.xlsx' with the path to your Excel file
file_path = "C://Users//jirip//Downloads//pohyb23//CZ0805.xlsx"
# Directory containing the Excel files
directory_path = "C://Users//jirip//Downloads//pohyb23//"

# Initialize an empty DataFrame to store results
all_data = pd.DataFrame()

# Loop through all files in the directory
for file_name in os.listdir(directory_path):
    if file_name.endswith(".xlsx"):
        file_path = os.path.join(directory_path, file_name)
        print(f"Processing file: {file_path}")
        df = pd.read_excel(file_path)
        columns_of_interest = df.iloc[:, [0, 1, 13]]
        columns_of_interest = columns_of_interest[columns_of_interest['Rok'] == 2023]
        print(f"Extracted length: {len(columns_of_interest)}")
        all_data = pd.concat([all_data, columns_of_interest], ignore_index=True)

columns_of_interest = all_data

# Display the extracted columns
print(columns_of_interest.head())

columns_of_interest.to_csv("C://Users//jirip//Downloads//pohyb23//all.csv", index=False)
```
